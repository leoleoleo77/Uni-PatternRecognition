```python
array = [0, 0, 0 ,0 ,0]
ocean_proximity_data = [0, 0, 0 ,0, 0]
num_of_data = 0
tempdata = dataMethods.Ocean_proximity_column()
for i in range(5):
    ocean_proximity_data[i] = 1
    for data in tempdata:
        if data == ocean_proximity_data:
            num_of_data += 1
    array[i] = num_of_data
    num_of_data = 0
    ocean_proximity_data[i] = 0
data = {'NEAR BAY': array[0], '<1H OCEAN': array[1], 'INLAND': array[2], 'NEAR OCEAN': array[3] ,'ISLAND': array[4]}
names = list(data.keys())
values = list(data.values())
fig, axs = plt.subplots(figsize=(9, 50))
axs.set_title('Ιστογράμματα συχνοτήτων των τιμών median ocean_proximity')
axs.bar(names, values)
```

```python
# set the size of the graph
fig, axs = plt.subplots(figsize=(10, 5))

# median house value stuff
array1 = dataMethods.Median_house_value_column()
array1_x_values = np.arange(0, 0.995, 0.005)
array1_y_values = axs.hist(array1, np.arange(0, 1, 0.005), density=True)[0]

# median house age stuff
array2 = dataMethods.Housing_median_age_column()
array2_x_values = np.arange(0, 0.98, 0.02)
array2_y_values = axs.hist(array2, np.arange(0, 1, 0.02), density=True)[0]

# total rooms stuff
array3 = dataMethods.Median_income_column()
array3_x_values = np.arange(0, 0.995, 0.005)
array3_y_values = axs.hist(array3, np.arange(0, 1, 0.005), density=True)[0]

# plt.close() so that the histograms created to get the y values wont show
plt.close()
# Plotting the specified values
plt.plot(array1_x_values, array1_y_values, marker='o', markersize=2, color='green', label="Median House Value")  
plt.plot(array2_x_values, array2_y_values, marker='o', markersize=2, color='orange', label="Median House Age")  
plt.plot(array3_x_values, array3_y_values, marker='o', markersize=2, color='purple', label="Median Income")  
# the the origin of the axis to be (0,0)
plt.xlim([0, 1]) 
plt.ylim([0, max(array3_y_values)])
plt.legend()
plt.title('Relation between the median house value, the median house age and the median income')
plt.grid(True)
plt.show()
```

```python
# set the size of the graph
fig, axs = plt.subplots(figsize=(10, 5))

# population stuff
array1 = dataMethods.Population_column()
array1_x_values = np.arange(0, 0.995, 0.005)
array1_y_values = axs.hist(array1, np.arange(0, 1, 0.005), density=True)[0]

# total rooms stuff
array2 = dataMethods.Total_rooms_column()
array2_x_values = np.arange(0, 0.995, 0.005)
array2_y_values = axs.hist(array2, np.arange(0, 1, 0.005), density=True)[0]

# total bed rooms stuff
array3 = dataMethods.Total_bedrooms_column()
array3_x_values = np.arange(0, 0.995, 0.005)
array3_y_values = axs.hist(array3, np.arange(0, 1, 0.005), density=True)[0]

# total households stuff
array4 = dataMethods.Households_column()
array4_x_values = np.arange(0, 0.995, 0.005)
array4_y_values = axs.hist(array4, np.arange(0, 1, 0.005), density=True)[0]

# plt.close() so that the histograms created to get the y values wont show
plt.close()
# Plotting the specified values
plt.plot(array1_x_values, array1_y_values, marker='o', markersize=2, color='green', label="Population")  
plt.plot(array2_x_values, array2_y_values, marker='o', markersize=2, color='red', label="Total Rooms")  
plt.plot(array3_x_values, array3_y_values, marker='o', markersize=2, color='blue', label="Total Bed Rooms")  
plt.plot(array4_x_values, array4_y_values, marker='o', markersize=2, color='orange', label="Households")  
# the the origin of the axis to be (0,0)
plt.xlim([0, 1]) 
plt.ylim(ymin=0.0)
plt.legend()
plt.title('Relation between the population, median rooms, the median bed rooms and the households')
plt.grid(True)
plt.show()
```
