```python
array = [0, 0, 0 ,0 ,0]
ocean_proximity_data = [0, 0, 0 ,0, 0]
num_of_data = 0
tempdata = dataMethods.Ocean_proximity_column()
for i in range(5):
    ocean_proximity_data[i] = 1
    for data in tempdata:
        if data == [1, 0, 0 ,0, 0]:
            num_of_data += 1
    array[i] = num_of_data
    num_of_data = 0
    ocean_proximity_data[i] = 0
data = {'NEAR BAY': array[0], '<1H OCEAN': array[1], 'INLAND': array[2], 'NEAR OCEAN': array[3] ,'ISLAND': array[4]}
names = list(data.keys())
values = list(data.values())
fig, axs = plt.subplots(figsize=(9, 3))
axs.set_title('Ιστογράμματα συχνοτήτων των τιμών median ocean_proximity')
axs.bar(names, values)

```
